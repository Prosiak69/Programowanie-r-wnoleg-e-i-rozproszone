```
from math import sin

def funkcja(my_func, a, b, n):
    h = (b-a)/n
    total = my_func(a)/2 + my_func(b)/2
    for i in range(1,n):
        x = a + h * i
        total += my_func(x)
    return total * h

wynik = funkcja(lambda x: sin(1.1*x-0.3)/(2.5+x*x), 0.4, 2.2, 10)


print(wynik)
```
