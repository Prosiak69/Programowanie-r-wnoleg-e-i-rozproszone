```
from math import sqrt

def x(i, a, b, n):
    return a+(i/n)*(b-a)

def t(i,a,b,n):
    return (x(i+1,a,b,n) + x(i,a,b,n)) / 2

def funkcja(my_func, a, b, n):
    h = (x(2,a,b,n) - x(1,a,b,n))/2
    total = my_func(x(0,a,b,n)) + my_func(x(n,a,b,n)) + 4*my_func(t(0,a,b,n))
    for i in range(1,n):
        total += 4*my_func(t(i,a,b,n))
        total += 2*my_func(x(i,a,b,n))
    return total * h/3

#wynik = funkcja(lambda x: sin(1.1*x-0.3)/(2.5+x*x), 0.4, 2.2, 10)

wynik = funkcja(lambda x: 0.5*(3*x*x-1)*sqrt(4*x*x+3), -1, 1, 1)


print(wynik)
```
